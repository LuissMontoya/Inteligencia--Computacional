# Histograma

```python
hist = cv2.calcHist(gray,[0],None,[256],[0,256])
```

```python
plt.plot(hist,color='red')
```

```python
color = ('r','g','b')
for i, c in enumerate(color):
    hist = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(hist,color=c)
    plt.xlim([0,256])
    
```

```python
EqH = cv2.equalizeHist(gray)
plt.imshow(EqH,cmap='gray')
```

```python
hist = cv2.calcHist([EqH],[0],None,[256],[0,256])
plt.plot(hist,color='gray')
```

# Binarización -- cv2. threshold

thr: umbral 

imgB: Imagen Umbralizada

```python
thr, imgB = cv2.threshold(gray,150,255,cv2.THRESH_BINARY)
plt.imshow(imgB, CMAP='gray')
```

```python
thr, imgB = cv2.threshold(gray,225,150,cv2.THRESH_OTSU)
plt.imshow(imgB, CMAP='gray')
```
