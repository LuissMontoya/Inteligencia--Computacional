```python
from bs4 import BeautifulSoup
import urllib.request
```

```python
respuesta = urllib.request.urlopen('https://librefinanciero.com/')
```

```python
html = respuesta.read()
```

```python
soup = BeautifulSoup(html, 'html5lib')
```

```python
text  = soup.get_text(strip = True)
```

```python
print(text)
```


# Tokenizar


```python
from nltk.tokenize import word_tokenize
```

```python
#Extraer tokens
tokens =word_tokenize(text,"spanish")

#Limpieza: remover los signos de puntuación

tokens = [word.lower() for word in tokens if word.isalpha()]

print(tokens)
```

```python
#verificar Frecuencia de cada palabra
freq = nltk.FreqDist(tokens)
for palabra, valor in freq.items():
    print(str(palabra) + ':' + str(valor))
```

```python
#Visualizar tokens
import matplotlib.pyplot as plt
freq.plot(30, cumulative=False)
```

```python
#Eliminar palabras de Parada
from nltk.corpus import stopwords
```

```python
#copia de la lista de tokens

Tokens_limpios = tokens[:]

for token in tokens:
     if token in stopwords.words('spanish'):
        Tokens_limpios.remove(token)
```

```python
print(Tokens_limpios)
```

```python
#verificar Frecuencia de cada palabra
freq_Limpia = nltk.FreqDist(Tokens_limpios)
for palabra, valor in freq_Limpia.items():
    print(str(palabra) + ':' + str(valor))
```

```python
#visualizar tokens
freq_Limpia.plot(30, cumulative= False)
```

```python
#Obtener sinónimos/ Antónimos
from nltk.corpus import wordnet

sinonimos = []

for syn in wordnet.synsets('investment'):
    for lemma in syn.lemmas():
        sinonimos.append(lemma.name())
    
print(sinonimos)
```

```python
#Reemplazar tokens sinónimos
sinonimos = ['libertad','independencia']
for ind, sin in enumerate(sinonimos):
    Sin_tokens_limpios = [word.replace(sinonimos[ind],'investment') for word in Tokens_limpios]
```

```python
#Recalcular frecuencia de cada palabra
freq_Limpia_sin = nltk.FreqDist(Sin_tokens_limpios)
for palabra, valor in freq_Limpia_sin.items():
    print(str(palabra) + ':' + str(valor))
```

```python
#visualizar tokens
freq_Limpia_sin.plot(30, cumulative= False)
```
